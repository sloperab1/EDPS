```python
import sympy 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = sympy.Symbol("x")
y = sympy.Symbol("y")
i = x**2 - y**2
j = 2*x*y
k=0
didx = sympy.diff(i,x)
djdy = sympy.diff(j,y)
div = didx+djdy
div
fig = plt.figure()
plot = fig.gca(projection='3d')
a,b,c = np.meshgrid(np.arange(-1,1,.2),np.arange(-1,1,.2),np.arange(-1,1,.2))
i,j,k = [a**2-b**2,2*a*b,0]
plot.quiver(a, b, c, i,j,k,length = .15)
plt.show()

```

```python
x,y = np.meshgrid(np.linspace(-20,20,20),np.linspace(-20,20,20))

i,j= [x**2-y**2,2*x*y]
plt.quiver(x,y,i,j)
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

u = -y/np.sqrt(x**2 + y**2)
v = x/np.sqrt(x**2 + y**2)

plt.quiver(x,y,u,v)
plt.show()
```
